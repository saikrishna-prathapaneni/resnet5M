```python
import matplotlib.pyplot as plt
```

```python
def loadData(path):
    arr=[]
    values=[]
    with open(path, 'r') as file:
        arr = file.read().split("\n")
    for x in arr:
        try:
            values.append(float(x))
        except:
            pass
    return values
```

## Comparing model architectures

```python
new_res=loadData("trainint_outputs/res_block/test_acc_history.txt")
res_512=loadData("trainint_outputs/res_512/test_acc_history.txt")
res_512_d=loadData("trainint_outputs/res_512_d/test_acc_history.txt")
og_model=loadData("trainint_outputs/og_resnet/test_acc_history.txt")
og_64_resenet=loadData("trainint_outputs/og_64_resnet/test_acc_history.txt")

x=range(0,200)
plt.plot(x, new_res, label='new_res')
plt.plot(x, res_512, label='res_512')
plt.plot(x, res_512_d, label='res_512_d')
plt.plot(x, og_model, label='og_model')
plt.legend()
plt.show()
plt.show()
```

```python
new_res=loadData("trainint_outputs/new_res/test_loss_history.txt")
res_512=loadData("trainint_outputs/res_512/test_loss_history.txt")
res_512_d=loadData("trainint_outputs/res_512_d/test_loss_history.txt")
og_model=loadData("trainint_outputs/og_model/test_loss_history.txt")

x=range(0,200)
plt.plot(x, new_res, label='new_res')
plt.plot(x, res_512, label='res_512')
plt.plot(x, res_512_d, label='res_512_d')
plt.plot(x, og_model, label='og_model')
plt.legend()
plt.show()
plt.show()
```

## Comparing Optimisers

```python
adadelta=loadData('./trainint_outputs/changing_optimisers/adadelta/test_acc_history.txt')
adam=loadData('./trainint_outputs/changing_optimisers/adam/test_acc_history.txt')
rmsprop=loadData('./trainint_outputs/changing_optimisers/rmsprop/test_acc_history.txt')
sgd=loadData('./trainint_outputs/changing_optimisers/sgd/test_acc_history.txt')

x=range(0,50)
plt.plot(x, adadelta, label='AdaDelta')
plt.plot(x, adam, label='Adam')
plt.plot(x, rmsprop, label='RMS Prop')
plt.plot(x, sgd, label='SGD')
plt.legend()
plt.show()
plt.show()
```

```python
gelu=loadData('trainint_outputs/changing_activations/gelu/test_acc_history.txt')
leakyrelu=loadData('trainint_outputs/changing_activations/leakyrelu/test_acc_history.txt')
tanh=loadData('trainint_outputs/changing_activations/tanh/test_acc_history.txt')
relu=loadData('trainint_outputs/changing_optimisers/sgd/test_acc_history.txt')

x=range(0,50)
plt.plot(x, gelu, label='GELU')
plt.plot(x, leakyrelu, label='LeakyReLU')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, relu, label='Relu')
plt.legend()
plt.show()
plt.show()
```
