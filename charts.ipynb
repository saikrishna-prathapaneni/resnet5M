```python
import matplotlib.pyplot as plt
```

```python
def loadData(path):
    arr=[]
    values=[]
    with open(path, 'r') as file:
        arr = file.read().split("\n")
    for x in arr:
        try:
            values.append(float(x))
        except:
            pass
    return values
```

## Comparing model architectures

```python
new_res=loadData("training/res_block/test_acc_history.txt")
res_512=loadData("training/res_512/test_acc_history.txt")
res_512_d=loadData("training/res_512_d/test_acc_history.txt")
og_model=loadData("training/og_resnet/test_acc_history.txt")
og_64_resenet=loadData("training/og_64_resnet/test_acc_history.txt")

x=range(0,200)
plt.plot(x, new_res, label='res_block')
plt.plot(x, res_512, label='res_512')
plt.plot(x, res_512_d, label='res_512_d')
plt.plot(x, og_model, label='og_resnet')
plt.plot(x, og_64_resenet, label='og_64_resnet')
plt.legend()
plt.title("test accuracy")
plt.show()
```

```python
new_res=loadData("training/res_block/test_loss_history.txt")
res_512=loadData("training/res_512/test_loss_history.txt")
res_512_d=loadData("training/res_512_d/test_loss_history.txt")
og_model=loadData("training/og_resnet/test_loss_history.txt")
og_64_resenet=loadData("training/og_64_resnet/test_loss_history.txt")

x=range(0,200)
plt.plot(x, new_res, label='res_block')
plt.plot(x, res_512, label='res_512')
plt.plot(x, res_512_d, label='res_512_d')
plt.plot(x, og_model, label='og_resnet')
plt.plot(x, og_64_resenet, label='og_64_resnet')
plt.legend()
plt.title("test loss")
plt.show()
```

```python
res_512=loadData("training/res_512/train_acc_history.txt")
res_512_d=loadData("training/res_512_d/train_acc_history.txt")
res_512_test=loadData("training/res_512/test_acc_history.txt")
res_512_d_test=loadData("training/res_512_d/test_acc_history.txt")


x=range(0,200)
plt.plot(x, res_512, label='res_512 - train acc')
plt.plot(x, res_512_d, label='res_512_d - train acc')
plt.plot(x, res_512_test, label='res_512 - test acc')
plt.plot(x, res_512_d_test, label='res_512_d - test acc')
plt.legend()
plt.title("train accuracy")
plt.show()
```

## Comparing Optimisers

```python
adadelta=loadData('./training/changing_optimisers/adadelta/test_acc_history.txt')
adam=loadData('./training/changing_optimisers/adam/test_acc_history.txt')
rmsprop=loadData('./training/changing_optimisers/rmsprop/test_acc_history.txt')
sgd=loadData('./training/changing_optimisers/sgd/test_acc_history.txt')

x=range(0,50)
plt.plot(x, adadelta, label='AdaDelta')
plt.plot(x, adam, label='Adam')
plt.plot(x, rmsprop, label='RMS Prop')
plt.plot(x, sgd, label='SGD')
plt.legend()
plt.show()
plt.show()
```

```python
gelu=loadData('training/changing_activations/gelu/test_acc_history.txt')
leakyrelu=loadData('training/changing_activations/leakyrelu/test_acc_history.txt')
tanh=loadData('training/changing_activations/tanh/test_acc_history.txt')
relu=loadData('training/changing_optimisers/sgd/test_acc_history.txt')

x=range(0,50)
plt.plot(x, gelu, label='GELU')
plt.plot(x, leakyrelu, label='LeakyReLU')
plt.plot(x, tanh, label='Tanh')
plt.plot(x, relu, label='Relu')
plt.legend()
plt.show()
plt.show()
```
